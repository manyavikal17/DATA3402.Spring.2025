1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class matrix:
    def __init__(self, n=None, m=None, values=None):     ## Matrix with either dimension (n,m) or list of lists(values)
    
        if values:   ## All rows have same number of cols
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError("All rows must have same number of columns")
            self.matrix =  values
            self.n = len(values)  
            self.m = len(values[0])
    
        elif n is not None and m is not None:  ## Creating a n*m zero matrix
            self.n = n
            self.m = m
            self.matrix = [[0] * m for _ in range(n)]
        
        else: 
            raise ValueError("Give either dimensions (n,m) or list of lists")
        
    def __repr__(self):
        return "\n".join([str(row) for row in self.matrix])
```

2. Add the following methods:
    * `shape()`: returns a tuple `(n,m)` of the shape of the matrix.
    * `transpose()`: returns a new matrix instance which is the transpose of the matrix.
    * `row(n)` and `column(n)`: that return the nth row or column of the matrix M as a new appropriately shaped matrix object.
    * `to_list()`: which returns the matrix as a list of lists.
    *  `block(n_0,n_1,m_0,m_1)` that returns a smaller matrix located at the n_0 to n_1 columns and m_0 to m_1 rows. 
    * (Extra credit) Modify `__getitem__` implemented above to support slicing.
        

```python
class matrix:
    def __init__(self, n=None, m=None, values=None):
    
        if values:
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError("All rows must have same number of columns")
            self.matrix =  values
            self.n = len(values)
            self.m = len(values[0])
    
        elif n is not None and m is not None:
            self.n = n
            self.m = m
            self.matrix = [[0] * m for _ in range(n)]
        
        else: 
            raise ValueError("Give either dimensions (n,m) or list of lists")
        
    def __repr__(self):
        return "\n".join([str(row) for row in self.matrix])
    
    def shape(self):    
        return (self.n, self.m)    ## Returning shape of matrix as a tuple
    
    def transpose(self):
        transpose = [[self.matrix[j][i] for j in range(self.n)] for i in range(self.m)]
        return matrix(values = transpose)
    
    def row(self, n):
        if not (0 <= n < self.n):   ## Returning a specific row
            raise IndexError("Error, Row index is out of range")
        return matrix(values = [self.matrix[n]])
        
    def column(self, m):        ## Returning a specific col
        if not (0 <= m < self.m):
            raise IndexError("Error, Column index is out of range")
        return matrix(values = [[self.matrix[i][m]] for i in range(self.n)])
        
    def to_list(self):
        return self.matrix      ## Returning matrix as list of lists

    def block(self, n_0, n_1, m_0, m_1):    ## Returning a small part of matrix
        if not(0 <= n_0 < n_1 <= self.n and 0 <= m_0 < m_1 <= self.m):
            raise ValueError("Invalid entry")
        smaller_matrix = [row[m_0 : m_1] for row in self.matrix[n_0 : n_1]]
        return matrix(values = smaller_matrix)
    
    def __getitem__(self, index):    ## Slicing matrix
        if isinstance(index, tuple):
            row , col = index
            if isinstance(row, slice) or isinstance(col, slice):
                smaller_matrix = [r[col] for r in self.matrix[row]]
                return matrix(values = smaller_matrix)
            return self.matrix[row][col]
        return self.matrix[index]
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m ,c):    ## Returning matrix filled with floats of c
    return matrix(values= [[float(c)] * m for _ in range(n)])

def zeros(n, m):          ## Returning matrix filled with floats of 0
    return matrix(values= [[float(0)] * m for _ in range(n)])

def ones(n, m):           ## Returning matrix filled with floats of 1
    return matrix(values= [[float(1)] * m for _ in range(n)])

def eye(n):               ## Returning an identity matrix 
    return matrix(values= [[1.0 if i==j else 0.0 for j in range(n)] for i in range(n)])

```

4. Add the following member functions to your class. Make sure to appropriately test the dimensions of the matrices to make sure the operations are correct.
    * `M.scalarmul(c)`: a matrix that is scalar product $cM$, where every element of $M$ is multiplied by $c$.
    * `M.add(N)`: adds two matrices $M$ and $N$. Donâ€™t forget to test that the sizes of the matrices are compatible for this and all other operations.
    * `M.sub(N)`: subtracts two matrices $M$ and $N$.
    * `M.mat_mult(N)`: returns a matrix that is the matrix product of two matrices $M$ and $N$.
    * `M.element_mult(N)`: returns a matrix that is the element-wise product of two matrices $M$ and $N$.
    * `M.equals(N)`: returns true/false if $M==N$.

```python
class matrix:
    def __init__(self, n=None, m=None, values=None):
    
        if values:
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError("All rows must have same number of columns")
            self.matrix =  values
            self.n = len(values)
            self.m = len(values[0])
    
        elif n is not None and m is not None:
            self.n = n
            self.m = m
            self.matrix = [[0] * m for _ in range(n)]
        
        else: 
            raise ValueError("Give either dimensions (n,m) or list of lists")
        
    def __repr__(self):
        return "\n".join([str(row) for row in self.matrix])
    
    def shape(self):
        return (self.n, self.m)
    
    def scalarmul(self, c):     ## Multiplies each element with scalar c
        return matrix(values= [[c * self.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def add(self, N):
        if self.shape() != N.shape():    ## Adds two matrices element-wise
            raise ValueError("Matrices must have equal dimensions for addition")
        return matrix(values= [[self.matrix[i][j] + N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def sub(self, N):
        if self.shape() != N.shape():    ## Subtracts two matrices element-wise
            raise ValueError("Matrices must have equal dimensions for subtraction")
        return matrix(values= [[self.matrix[i][j] - N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def mat_mult(self, N):
        if self.m != N.n:
            raise ValueError("Matrix multiplication not possible")
        result = [[sum(self.matrix[i][k] * N.matrix[k][j] for k in range(self.m)) for j in range(N.m)] for i in range(self.n)]
        return matrix(values= result)
    
    def element_mult(self, N):
        if self.shape() != N.shape():     ## Performs Matrix Multiplication
            raise ValueError("Matrices must have equal dimensions for multiplication")
        return matrix(values= [[self.matrix[i][j] * N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def equals(self, N):          ## Checks if two matrices are equal
        if self.shape() != N.shape():
            return False
        return all(self.matrix[i][j] == N.matrix[i][j] for i in range(self.n) for j in range(self.m))
```

5. Overload python operators to appropriately use your functions in 4 and allow expressions like:
    * 2*M
    * M*2
    * M+N
    * M-N
    * M*N
    * M==N
    * M=N


```python
class matrix:
    def __init__(self, n=None, m=None, values=None):
    
        if values:
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError("All rows must have same number of columns")
            self.matrix =  values
            self.n = len(values)
            self.m = len(values[0])
    
        elif n is not None and m is not None:
            self.n = n
            self.m = m
            self.matrix = [[0] * m for _ in range(n)]
        
        else: 
            raise ValueError("Give either dimensions (n,m) or list of lists")
        
    def __repr__(self):
        return "\n".join([str(row) for row in self.matrix])
    
    def shape(self):
        return (self.n, self.m)
    
    def scalarmul(self, c):
        return matrix(values= [[c * self.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def add(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrices must have equal dimensions for addition")
        return matrix(values= [[self.matrix[i][j] + N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def sub(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrices must have equal dimensions for subtraction")
        return matrix(values= [[self.matrix[i][j] - N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def mat_mult(self, N):
        if self.m != N.n:
            raise ValueError("Matrix multiplication not possible")
        result = [[sum(self.matrix[i][k] * N.matrix[k][j] for k in range(N.n)) for j in range(N.m)] for i in range(self.n)]
        return matrix(values= result)
    
    def element_mult(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrices must have equal dimensions for multiplication")
        return matrix(values= [[self.matrix[i][j] * N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def equals(self, N):
        if self.shape() != N.shape():
            return False
        return all(self.matrix[i][j] == N.matrix[i][j] for i in range(self.n) for j in range(self.m))
    
    def __mul__(self, c):
        if isinstance(c, (int, float)):
            return self.scalarmul(c)
        elif isinstance(c, matrix):
            return self.mat_mult(c)
        else: 
            raise TypeError("Matrix can only be multiplied by a scalar or another matrix")
            
    def __rmul__(self, c):
        return self.__mul__(c)
    
    def __add__(self, c):
        if isinstance(c, matrix):
            return self.add(c)
        else:
            raise TypeError("Matrices must have equal dimensions for addition")
    
    def __sub__(self, c):
        if isinstance(c, matrix):
            return self.sub(c)
        else:
            raise TypeError("Matrices must have equal dimensions for subtraction")
            
    def __eq__(self, N):
        if not isinstance(N, matrix):
            return False
        return self.equals(N)

    
```

6. Demonstrate the basic properties of matrices with your matrix class by creating two 2 by 2 example matrices using your Matrix class and illustrating the following:

$$
(AB)C=A(BC)
$$
$$
A(B+C)=AB+AC
$$
$$
AB\neq BA
$$

$$

```python
A = matrix(values=[[1, 2], [3, 4]])
B = matrix(values=[[2, 5], [1, 2]])
C = matrix(values=[[1, 1], [2, 3]])
I = eye(2)

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nMatrix C:")
print(C)
print("\nIdentity Matrix:")
print(I)


# FIRST PROPERTY
AB = A * B
ABC1 = AB * C
BC = B * C
ABC2 = A * BC
print("\n(AB)C:")
print(ABC1)
print("\nA(BC):")
print(ABC2)
print("Are they equal?", ABC1 == ABC2)


# SECOND PROPERTY
B_plus_C = B + C
A_B_plus_C = A * B_plus_C  # A(B + C)
AC = A * C
A_B_plus_A_C = AB + AC   # AB + AC
print("\nA(B+C):")
print(A_B_plus_C)
print("\nAB + AC:")
print(A_B_plus_A_C)
print("Are they equal?", A_B_plus_C == A_B_plus_A_C)


# THIRD PROPERTY
BA = B * A
print("\nAB:")
print(AB)
print("\nBA:")
print(BA)
print("Are they equal?", AB == BA) 


# FOURTH PROPERTY
AI = A * I
print("\nAI:")
print(AI)
print("\nA:")
print(A)
print("Are they equal?", AI == A)
```
